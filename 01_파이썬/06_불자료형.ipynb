```python
print(1 == 1) # ==(같다는 뜻을 가진 비교연산자)

print(2 > 1) # > (컴퓨터는 왼쪽항 기준이다. 즉 왼쪽이 더 크다)

print(1 >= 2) # 왼쪽 항을 기준으로 크거나 같지 않음. =은 부등호 뒤에 온다.

print(1 != 2) # !=(1은 2와 다르다. 같지 않다.) 
```

```python
print(bool(1))

print(bool(0))

print(bool(""))

print(bool("python"))

print(bool([]))

print(bool([1, 2]))

# 값이 비어있으면 False / 값이 있으면 Ture로 인식
```

```python
# and 연산자
print(True and True)

print(True and False)

print(False and False)

# or 연산자

print(True or True)

print(True or False)

print(False or False)
```
