```python
s2 = set("hello")
s2

# 집합자료형은 중복된 자료는 제거함. 그리고 자동적으로 오름차순으로 정렬 됨
```

```python
s1 = set([1, 2, 3, 3, 3, 7, 10, 2, 105, 70, 90]) #예를 들어 이게 빅 데이터라고 봤을 때 중복된 값은 지워짐
l1 = list(s1) # 집합 형태의 Data를 리스트 형으로 바꿈 그리고 l1 변수로 지정함.
print(l1)
#가장 빠르게 중복된 자료를 지우고 리스트 형식으로 만드는 방법
```

```python
## 교집합, 합집합, 차집합 구하기
# 빅데이터 시험이랑 AICE 시험에서 많이 요구함
s1 = set([1, 2, 3, 4, 5, 6, 7, 8])
s2 = set([4, 5, 6, 7, 8, 9, 10, 11])
```

```python
# 교집합 구하기
print(s1 & s2) #둘의 교차되는 점을 구한다 할 때 &(And 연산자 사용)
```

```python
# 교집합 구하기 2
s1.intersection(s2)
```

```python
# 합집합 구하기
s1 | s2 # 합집합을 구할 때 |(or 연산자) 사용
```

```python
# 합집합 구하기 2
s1.union(s2)
```

```python
# 차집합 구하기
print(s1 - s2) # s1과 s2와 공통인 영역을 s1에서 뺀 값을 구함
print(s2 - s1)
```

```python
# 차집합 구하기
print(s1.difference(s2))
print(s2.difference(s1))
```
