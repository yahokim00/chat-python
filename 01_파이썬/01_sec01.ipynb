```python
### 비교연산자
```

```python
# 크다 >
# 크거나 같다(이상) >=
# 작다 <
# 작거나 같다(이하) <=
# 같다 ==
# 다르다 !=

# 코드는 위에서 아래로 그리고 왼쪽에서 오른쪽으로 감. 즉, 3 > 2 왼쪽 3이 오른쪽 2 보다 크다.
```

```python
# 참조건
print(10 > 5)
print(10 < 5)
print(10 <= 10)
print(10 == 10)
print(10 != 10)

# print(10 = 10) -> 이건 에러뜸 -> 컴퓨터에서는 =가 오른쪽을 왼쪽에다가 대입하겠다 이 말이기 때문에. 변수 만들 때는 숫자가 처음 올 수 없음
# 논리식으로 비교 연산자로 인해 true 아니면 flase 값이 나옴
```

```python
### 조건문
score = 85

if score >= 90:
    print("a")
else:
    print("b")

# : 의미는 then -> ~~ 이면 ~~ 해라 위의 경우 만약 score가 90보다 크거나 같으면 print a 해라 
```

```python
점수 = 87

if 점수 >= 90:
    print("a")
elif 점수 >= 80:
    print("b")
elif 점수 >= 70:
    print("c")
else:
    print("f")
# elif 점수 >= 80: 이랑 70이랑 순서 바꾸면 c 만 나오고 끝. 즉, 위에서 아래로 만족하는 조건 하나가 나오면 그에 맞는 어떤 것을 하고 끝낸다
```

```python
score = 2

if score >= 10:
    print("score가 10 이상입니다.")
elif score < 10 and score >= 5:
    print("score가 5이상 10 미만입니다.")
elif score < 5 and score >= 3:
    print("score가 3이상 5 미만입니다.")
else:
    print("score가 3미만입니다.")
```
