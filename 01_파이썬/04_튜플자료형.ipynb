```python
## 튜플은 리스트 형태와 비슷
## 리스트는 수정이 가능하고, 튜플은 수정이 불가능
```

```python
# 튜플 생성
# 튜플은 하나의 요소더라도 값을 적고 ,를 적어야함
t1 = ()
t2 = (1,) #l2 = [1]
t3 = (1, 2, 3)
t4 = 1, 2, 3 # 튜플에서는 괄호 생략됨 -> t3이랑 t4랑 같음

# esult = 108, "많음"
t5 = ("a", "b", (1, 2))
```

```python
# 튜플 요소를 삭제

# 리스트요소 예시
list1 = [1, 2, 3]
del list1[1]
print("리스트요소삭제:", list1)

# t1 = (1, 2, 3)
# del t1[1]
# print("튜플요소삭제:", t1) -> 이건 삭제 안돼서 에러 뜸 
```

```python
# 리스트 요소를 수정
list1 = [1, 2, 3]
list1[1] = 4
print(list1)

# 튜플 요소를 수정
# t1 = (1, 2, 3)
# t1[1] = 4
# print("튜플요소를 수정:", t1) -> 동일하게 에러뜸 -> 즉, 튜플은 수정 및 새롭게 재정의도 안됨
```
