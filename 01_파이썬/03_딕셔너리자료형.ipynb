```python
# 딕셔너리 생성

a = {1: "a"}

# 딕셔너리 쌍 추가
a[2] = "b"

print(a)
```

```python
a["name"] = "pey"
```

```python
a[3] = [1, 2, 3]
print(a)
```

```python
ken = {"name" : "홍길동", "gender": "male", "hobby": "running"}
print(ken["name"])
print(ken["gender"])
print(ken["hobby"])
```

```python
listbox = ken.keys()
result = list(listbox)
result
```

```python
list_values = ken.values()
result = list(list_values)
print(result)
```

```python
# ken["name"]
ken.get("name")
```

```python
ken
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
print("name" in ken)
print("email" in ken)
```

```python
# key로 value 값 얻기 = pop
ken = {'name': '홍길동', 'gender': 'male', 'hobby': 'running'}
result = ken.pop("hobby")
print(result)
```
