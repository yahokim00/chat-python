```python
### while 조건문의 기본 구조

# while 조건문:
#     수행할 문장 1 
#     수행할 문장 2
```

```python
treehit = 0
while treehit < 10:
    treehit = treehit + 1
    print("나무를 %d번 찍었습니다." % treehit) # % 에 treehit의 변수를 집어넣는거임
    if treehit == 10:
        print("나무 넘어갑니다.")
# 10 될 때 까지 계속 반복하다가 마지막 문장에서 조건을 만족할 때 끝나는게 while 구문
```

```python
prompt = """
1. 추가
2. 삭제
3. 조회
4. 끝

숫자를 입력하세요:
"""

number = 0
while number !=4:
    print(prompt)
    number = int(input())


# a = "사과
# 바나나"
# 따옴표 2개는 다음 줄로 못넘어가는데 따옴표 3개 하면 줄 바꿈이 있는 문자로 인식함
```

```python
coffee = 10
money = 300
while money: 
    print("돈을 받았으니 커피를 줍니다")
    coffee = coffee - 1
    print("남은 커피의 양은 %d개 입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다")
        break # while문에서는 break 적어놓으면 프로그램 종료함
```

```python
coffee = 10
while True:
    money = int(input("돈을 넣어주세요")) # int는 정수로 만들어주는 함수 #input으로 가져오는 데이터는 문자임 정수가 아님 그럼 if 구문에서 바로 Flase 날거임 숫자가 아니니까

    if money == 300:
        print("커피를 줍니다")
        coffee = coffee - 1
    elif money > 300:
        print("거스름톤 %d를 주고 커피를 줍니다." % (money - 300))
        coffee = coffee - 1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다")
        print("남은 커피의 개수는 %d개입니다." % coffee )
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다")
        break
```
