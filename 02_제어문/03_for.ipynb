```python
### for 문의 기본구조
# for 변수 in 리스트(또는 튜플, 문자열):
#   수행할 문장1
#   수행할 문장2
#   ...
```

```python
li = ['one', 'two', 'three']
for i in li:
    print(i)
```

```python
a = [(1, 2), (3, 4), (5, 6)] #a는 3개의 요소가 있으니 3번 반복한다 a의 요소의 수만큼 구문을 반복한다.
for first, last in a: # ()가 있는 것과 같음
    print(first + last)

(first, last) = (1, 2)
```

```python
scores = [95, 70, 60, 55, 75, 97, 88]
number = 0

for score in scores:
    number = number + 1
    if score >= 70:
        print(f"{number}번 학생은 합격입니다.")
    else:
        print(f"{number}번 학생은 불합격입니다.")
# 반복문을 쓸 때는 for을 많이 씀
```

```python
# for 문과 함께 자주 사용하는 range 함수
a = range(10)
print(list(a))
```

```python
a = range(1 ,11)
print(list(a))
```

```python
# 1부터 100까지 더하기
result = 0

for i in range(1,100+1):
    result = result + i ## 오른쪽항부터 계산

print(result)
```

```python
for i in range(1, 3):
    for j in range(1, 3):
        print(i + j)
```

```python
# 구구단을 만드는데
# 출력값을
#2 * 1 = 2
#2 * 2 = 4
#2 * 3 = 6

for first in range(2, 9+1):
    for second in range(1, 9+1):
        print(f"{first} * {second} = {first*second}")
```
