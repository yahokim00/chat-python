```python
money = 1000

if money:
    print("택시를 타고 가라")
else:
    print("걸어가라")

# if 다음에는 조건문(logical Test) 논리식을 적어줘야함 -> 참과 거짓이 필요한데 근데 머니를 적음 -> 머니는 숫자 1000으로 숫자는 0이 아니면 참이니 택시를 타고 가라를 출력
# if이면 위에꺼부터 참이면 실행하고 거짓이면 else 실행
# : 이건 이 문장은 이 줄에서는 이게 끝이고 나머지 내용은 밑에서 들여쓰기 해야함
```

```python
money = 2000

if money >= 3000:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
money = 2000
card = True

if money >= 3000 or card:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
## in, not in 
# x in 리스트
# x in 튜플
# x in 문자열

print(1 in [1, 2, 3]) # 1이 리스트 안에 있는지 없는지 -> 있으니까 True
print("a" in ("a", "b", "c")) # a가 튜플 안에 있는지 없는지 -> 있으니까 True
print("j" in "python") # j가 문자열 안에 있는지 없는지 -> 없으니까 False

print(1 not in [1, 2, 3]) # 1이 리스트 안에 있는지 없는지 -> 있으니까 False
```

```python
pocket = ['paper', 'cellphone', 'money']

if 'money' in pocket:
    print("택시를 타고 가라")
else:
    print("걸어가라")
# money 값이 pocket 안에 있으니 택시를 타고 가라 출력
```

```python
pocket = ['paper', 'cellphone', 'money']

if 'money' in pocket:
    pass
else:
    print("카드를 꺼내라")
# money 값이 pocket 안에 있으니 그냥 패스라서 아무 값도 안나옴
# 만약 money가 아니라 다른 값이라서 pocket 안에 없으면 카드를 꺼내라 출력
```

```python
# 다양한 조건을 판단하는 elif
pocket = ['paper', 'cellphone']
card = True

if 'money' in pocket:
    print("택시를 타고 가라")
else:
    if card:
        print("택시를 타고 가라")
    else:
        print("걸어가라")

# money는 없으니까 else로 가서 card가 있으니까 택시를 타고 가라 출력
# 만약 card도 없었으면 걸어가라 출력 됨
```

```python
pocket = ['paper', 'cellphone']
card = False

if 'money' in pocket:
    print("택시를 타고 가라")
elif card:
    print("택시를 타고 가라")
else:
    print("걸어가라")

#위에 구문이랑 똑같은건데 다른 버전 느낌
```

```python
# 홍길동의 시험 점수
score = 80

# 컴퓨터 시험을 쳤는데 90점 이상이면 "a" 80점 이상이면 "b" 70점 이상이면 "c" 나머지는 "f"

if score >= 90:
    print("a")
elif score >= 80:
    print("b")
elif score >= 70:
    print("c")
else:
    print("f")    
```

```python
# 변수 = 참일 때 값 if 조건 else 거짓일 때 값
# 간단하게 한줄로 표현할 때
# 많이 쓰이진 않는데 간혹 쓰일 때 있음

age = 19
status = "성인" if age >= 18 else "미성년"
print(status)

# 아래 내용을 한 줄로 줄인게 위 내용

if age >= 19:
    print("성인")
else:
    print("미성년")
```
