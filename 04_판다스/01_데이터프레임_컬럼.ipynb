```python
# 03. 탐색적 데이터 분석(EDA)

# 어떤 데이터를 가져왔을 때 어떤 타입인지 분석해야함

df = pd.read_csv("temp.csv")

print(df.head()) # df.head하면 상위 5개 행만 가져옴 -> 수천개 행이 있을 때 먼저 조금 볼려고 
print(df.head(2)) # 이러면 2개만 가져옴.

print(df.tail()) # 하위 5개
print(df.tail(1)) # 마지막 값
```

```python
# 컬럼별 자료형
df.info() # Dtype은 먼 type 인지 알려줌
# 나중에 머신러닝 돌려야되는데 object 형태의 데이터 타입은 머신러닝 못돌림 -> 인코딩(숫자와 연결시켜주는) 해야함
# shape나 info는 시험에 나옴
```

```python
# 4. 자료형 변환
data = {
    "메뉴": ["아메리카노", "카페라떼", "카페모카", "카푸치노", "에스프레소", "밀크티", "녹차"],
    "가격": [1500.0, 2000.0, 3000.0, 2000.0, 1500.0, 3000.0, 4000.0],
    "칼로리": [10, 110, 250, 110, 20, 210, 0]
}

df = pd.DataFrame(data)
df.info()
```

```python
df["가격"] = df["가격"].astype(int)
```
