```python
import pandas as pd
df = pd.read_csv("cafe3.csv")
df
```

```python
# replace() 활용

df.replace("아메리카노", "롱고", inplace=True)
```

```python
df.replace("녹차", "그린티", inplace=True)
df
```

```python
# 여러개의 값을 한꺼번에 변경
change = {
    "롱고" : "아메리카노",
    "그린티" : "녹차"
}
df.replace(change, inplace=True)
df
```

```python
# Loc / iloc 활용

df.loc[6, "원산지"] = "대한민국"
df
```

```python
df.loc[1:2, "이벤트가"] = 1000
df
```

```python
df["이벤트가"] = df["이벤트가"].astype(float)
df.info()
```

```python
df.to_csv("cafe4.csv", index=False)
```
