```python
import pandas as pd

df = pd.read_csv("cafe2.csv")
df
```

```python
# 결측치 확인

df.isnull() # True 3개가 값이 비어있다는 말임 
# False는 0과 같음 True는 1과 같음
```

```python
df.isnull().sum()
# 결측치를 구할 때 isnull().sum()하면 결측치가 몇개인지 알 수 있음
```

```python
# 결측치 채우기
df["원산지"].fillna("코스타리카", inplace=True)
df
```
