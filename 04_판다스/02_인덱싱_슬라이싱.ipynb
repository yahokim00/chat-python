```python
## 필터링

df=pd.read_csv("cafe2.csv")
df
```

```python
 cond = df["칼로리"] == 20
 df
```

```python
df[cond]
```

```python
df[df["원산지"] == "과테말라"]
```

```python
# 문제. 칼로리가 200 이하이면서 가격이 5000원 이상인 메뉴를 필터링 하세요.

cond = (df["칼로리"] <= 200) & (df["가격"] >=2000)
df[cond]
```

```python
# Not 연산자
cond = df["칼로리"] <50
df[~cond]
```

```python
# 복수 조건 필터링
cond1 = df["가격"] > 5000
cond2 = df["칼로리"] >100

df[cond1 | cond2]
```
