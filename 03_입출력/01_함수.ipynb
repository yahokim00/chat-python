```python
# 매개변수를 지정하여 호출하기
def sub(a, b):
    return a - b

result = sub(a=10, b=2)
print(result)
```

```python
result = sub(b=5, a=3)
print(result)
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 함수를 만들지?
def add_money(*args):
    result = 0
    for i in args:
        result = result + i
    return result

hap = add_money(1, 2, 3, 4, 5)
print(hap)
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i
    
    return result

c = add_mul("add", 10, 30, 40)
print(c)
```

```python
c = add_mul("mul", 5, 3, 2)
print(c)
```

```python
# 함수의 값은 언제나 하나이다.

def add_and_mul(a, b):
    return a+b, a*b

result = add_and_mul(10, 20)
print(result)
```

```python
def add_and_mul(a, b):
    return a+b
    return a*b # 리턴은 하나만 됨. -> 하나 이후로는 음영 처리되고 실행 안되고 종료

c = add_and_mul(10, 20)
print(c)
```
